# Understanding Data
I will use `.info` and `.head`to get the necessary knowledge, see how many missing values are there and see the relevant columns to use.

```python
#First load the data and read it using pd
aviation_df=pd.read_csv('./data/Aviation_Data.csv')
aviation_df.info()
```

```python
aviation_df.describe()
```

```python
aviation_df.head()
```

```python
aviation_df.drop_duplicates(subset=['Event.Id'], keep='first', inplace=True)
aviation_df.info()
```

```python
aviation_df['Injury.Severity'].head(10)
```

```python
aviation_df=aviation_df.dropna(subset=['Event.Id'])
aviation_df.info()
```

```python
wanted_columns=['Event.Id','Location','Country','Event.Date','Make','Model','Purpose.of.flight','Total.Fatal.Injuries','Total.Serious.Injuries',
                'Total.Minor.Injuries','Total.Uninjured']
relevant_df=aviation_df[wanted_columns]
relevant_df.info()
```

```python
relevant_df.loc[:,'Total.Fatal.Injuries']=relevant_df['Total.Fatal.Injuries'].fillna(0)
relevant_df.loc[:,'Total.Serious.Injuries']=relevant_df['Total.Serious.Injuries'].fillna(0)
relevant_df.loc[:,'Total.Minor.Injuries']=relevant_df['Total.Minor.Injuries'].fillna(0)
relevant_df.loc[:,'Total.Uninjured']=relevant_df['Total.Uninjured'].fillna(0)
relevant_df.info()
```

```python
relevant_df['Purpose.of.flight'].value_counts()
relevant_df.loc[:,'Purpose.of.flight']= relevant_df['Purpose.of.flight'].fillna('Unknown')
```

```python
relevant_df.loc[:,'Make']=relevant_df['Make'].fillna('Unknown')
```

```python
relevant_df.loc[:,'Model']=relevant_df['Model'].fillna('Unknown')
relevant_df.loc[:,'Make-Model']=relevant_df['Make']+" "+relevant_df['Model']

relevant_df.loc[:,'Location']=relevant_df['Location'].fillna('Unknown')

relevant_df.loc[:,'Country']=relevant_df['Country'].fillna(relevant_df['Country'].mode()[0])
relevant_df.info()
```

```python
def clean_string(s):
    if type(s) == str:
        s = s.strip()  # Remove leading and trailing spaces
        s = ''.join(e for e in s if e.isalnum() or e.isspace())  # Keep alphanumeric characters and spaces
    return s
```

```python
relevant_df=relevant_df.map(clean_string)
```

```python
relevant_df['Event.Date']=relevant_df['Event.Date'].astype(str)
relevant_df['Year']=relevant_df['Event.Date'].str[:4]
relevant_df['Month'] = pd.to_datetime(relevant_df['Event.Date'], format='%Y%m%d').dt.strftime('%B')

relevant_df.info()
```

```python
relevant_df['Total.Injured']=relevant_df[['Total.Fatal.Injuries', 'Total.Serious.Injuries', 'Total.Minor.Injuries']].sum(axis=1)
relevant_df['Total.Injured']
relevant_df.info()
```
